```python
import matplotlib.pyplot as plt
```

```python
#Line plot
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
```

```python
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line plot")
plt.plot(x,y)
```

```python
plt.plot(x,y, linestyle='dashed', color='green', marker='o', linewidth=2) 
```

```python
x=[1,2,3,4,5,6]
y=[2000,3000,6000,4000,3500,7000]
plt.xlabel("Months(1-6)")
plt.ylabel("Savings amount")
plt.title("Growth of savings amount over 6 months") 
plt.plot(x,y, linestyle='dashed', color='green', marker='o', linewidth=2) 
```

```python
#Bar plot
```

```python
cat=['A','B','C','D','E']
values=[5,9,3,12,6]
plt.bar(cat,values,color='brown')
plt.title("Bar chart")
plt.xlabel("Categories")
plt.ylabel("Values")
```

```python
plt.bar(cat,values,color=['yellow', 'purple','blue', 'red','green'])
```

```python
plt.bar(cat,values,color='grey', edgecolor='purple')
```

```python
#Histogram
```

```python
ages=[20,22,33,35,36,32,32,33,31,43,44,46,44,52,54,57,53,59,60,65,64,]
b=[20,30,40,50,60]
# bins: Defines the intervals for grouping data
plt.hist(ages,bins=b,color='skyblue',edgecolor='blue')
plt.title("Age distribution")
plt.xlabel("Age ranges")
plt.ylabel("Frequency")
plt.show()
```

```python
lang=['Python','Java','C++','JS']
percentage=[45,15,25,15] 
colors=['skyblue','pink','green','grey']
plt.pie(percentage,labels=lang,autopct='%1.1f%%',colors=colors)
plt.title("Languages")
plt.legend()
```

```python
x=[5,10,15,20,25]
y=[10,20,25,30,35]
plt.scatter(x,y, color='green', s=100, edgecolor='black')
plt.title("Scatter plot")
plt.xlabel('X values')
plt.ylabel('Y values')
```
