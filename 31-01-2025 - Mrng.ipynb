```python
from sklearn.preprocessing import LabelEncoder
data={'Study_level': ['Begginer','Intermediate','Advanced'],
     'level':[3,4,5]}
df=pd.DataFrame(data)
encoder = LabelEncoder()
df['Study_encoded']=encoder.fit_transform(df['Study_level'])
print(df)
correlation=df['Study_encoded'].corr(df['level'])
print("Correlation: ",correlation)
```

```python
data={'Cab_services':['Very Bad','Bad','Satisfied','Good','Excellent'],
     'rating':[1,2,3,4,5]}
df=pd.DataFrame(data)
df_encoded=pd.get_dummies(df,columns=['Cab_services'])
print(df_encoded)
correlation=df_encoded['Cab_services_Bad'].corr(df['rating'])
print("Correlation: ",correlation)

#get_dummies to get True or False
#encoder to get 0 or 1
```
